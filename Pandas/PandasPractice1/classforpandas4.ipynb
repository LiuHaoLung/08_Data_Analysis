```python
djson = pandas.read_json('/Users/How.Short/Desktop/Python內容/Python-class/classforpandas/Employees.json')
```

```python
djson.set_index('ID',inplace = True)
```

```python
djson
```

```python
djson[djson.loc[:,'Salary'] < 50000]
```

設定條件，salary 要小於 50000。

```python
djson[(djson.loc[:,'Salary'] < 50000) | (djson.loc[:,'Salary'] > 56000)]
```

要選出 salary 是介於 50000 和 56000 之間的值。

<code>|</code> 就是 or 。

```python
djson[(djson.loc[:,'Salary'] > 50000) & (djson.loc[:,'Department'] == 'Sales')]
```

要選出 salary 大於 50000，且部門是 sales。

<code>&</code>就是 and。

```python
djson[~(djson.loc[:,'Department'] == 'IT')]
```

回傳 department 不是 IT 的值。

<code>~</code>就是 not。
