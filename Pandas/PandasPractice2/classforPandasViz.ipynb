```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# 將 df1 這個 dataframe index 設定為第一行
df1 = pd.read_csv('df1',index_col = 0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
# 將要製作成的圖表欄位指定出來，並用 hist() 方法，就可以畫直條圖

df1['A'].hist()
```

```python
# 可以用 matplotlib 的 argument

df1['A'].hist(bins = 30)
```

```python
# 除了直接使用 hist()，也可以用 plot，並回傳 kind argument

df1['A'].plot(kind = 'hist')
```

```python
# 可以直接 .plot.hist() 來繪製直條圖

df1['A'].plot.hist()
```

```python
df2.head()
```

```python
df2.plot.area(alpha = 0.4)
```

```python
# index should be categorical

df2.plot.bar()
```

```python
# 將圖變成堆疊的方式

df2.plot.bar(stacked = True)
```

```python
# lineplot should be specift the x and y
# the default x is index

df1.plot.line(y='B')
```

```python
df1.plot.line(y='B', figsize = (12,3))
```

```python
df1.plot.line(y='B', figsize = (12,3), lw = 1)
```

```python
# scatterplot should be specift the x and y

df1.plot.scatter(x = 'A', y = 'B')
```

```python
# three dimensional plot
# c columns is displayed by color

df1.plot.scatter(x = 'A', y = 'B', c = 'C')
```

```python
# three dimensional plot
# c columns is displayed by color

df1.plot.scatter(x = 'A', y = 'B', c = 'C',cmap = 'coolwarm')
```

```python
# three dimensional plot
# large marker is C

df1.plot.scatter(x = 'A', y = 'B', s = df1['C']*100)
```

```python
# all the columns

df2.plot.box()
```
